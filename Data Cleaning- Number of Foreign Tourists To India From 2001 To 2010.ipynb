### Cleaning Data

```python
df.fillna(0, inplace=True) #Replace all NaN elements with 0s
```

```python
df.drop_duplicates() #Drop duplicates except for the first occurrence.
```

#### Replacing U.S.A by United States of America:

```python
#The replace() method searches a string for a specified value, or a regular expression, and returns a new string where the specified values are replaced. 
df['Nationality'] = df['Nationality'].str.replace('U.S.A.', 'United States of America') 
df
```

#### Replacing STATELESS to NaN:
    (My intention is to replace it by NaN and drop it later on)

```python
df['Nationality'].replace('STATELESS', 'NaN', inplace=True) #replace STATELESS by NaN
```

```python
df #display dataframe
```

```python
#Drop NaN values from specified column.
df.drop(df.loc[df['Nationality']=='NaN'].index, inplace=True)
```

```python
df #display dataframe
```

#### Replacing Nationality-All values of Others to NaN:
    (My intention is to replace all values associated with it by NaN and drop it later on)

```python
df[df['Nationality'].str.contains('Others')] = 'NaN' #replacing all values in Others by NaN
df
```

```python
df.drop(df.loc[df['Region']=='NaN'].index, inplace=True) #dropping the rows with Region NaN
df
```

#### Replacing Nationality-All values of Total to NaN:
    (My intention is to replace all values associated with it by NaN and drop it later on)

```python
df[df['Nationality'].str.contains('Total')] = 'NaN'  #replacing all values in Total by NaN
df
```

```python
df.drop(df.loc[df['Region']=='NaN'].index, inplace=True) #dropping the rows with Region NaN
df
```

#### Changing the case:

```python
#converts all uppercase characters in a string into lowercase characters and returns it.
df['Region'] = ( df['Region']).str.lower()
df
```

```python
#returns a copy of the original string and converts the first character of the string to a capital (uppercase) letter 
#while making all other characters in the string lowercase letters.
df['Region'] = ( df['Region']).str.capitalize()
df
```
